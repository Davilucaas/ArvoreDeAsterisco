```python
number = int(input("Qual o número da base da sua árvore?"))
lst = range(1, number+1)
n = 1
for i in lst:  
  if i == min(lst):
    print(" "*(max(lst)-i)+"*"*(i))
  elif i == max(lst):
    print(" "*(max(lst)-i)+"*"*(i+lst[i-2]))
  else:
    print(" "*(max(lst)-i)+"*"+" "*((i*2)-3)+"*")
```
